```python
import pandas as pd

# Load the dataset
data = pd.read_csv('Consumer_Complaints.csv')  # Update with the actual dataset file name

# Get unique products and issues
unique_products = data['Product'].unique()
unique_issues = data['Issue'].unique()

# Create a list to store the unique issues with their corresponding unique products and numbers
combined_table = []

# Assign a number to each unique product
product_numbers = {product: number for number, product in enumerate(unique_products, start=1)}

# Iterate through each unique issue
for issue in unique_issues:
    # Find the unique products associated with the current issue
    products = data[data['Issue'] == issue]['Product'].unique()
    
    # Add the issue, its associated products, and their numbers to the combined table
    combined_table.append({'Issue': issue, 'Products': ', '.join(products), 'Product Number': ', '.join(str(product_numbers[product]) for product in products)})

# Create a DataFrame from the combined table
combined_table_df = pd.DataFrame(combined_table)

# Save the DataFrame to a CSV file
combined_table_df.to_csv('combined_table_.csv', index=False)
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from deap import creator, base, tools, algorithms

# Load the preprocessed dataset
combined_table = pd.read_csv('combined_table.csv')  # Replace with the actual path to your dataset

# Split the data into features (X) and labels (y)
X = combined_table['Issue']
y = combined_table['Products']

# Vectorize the text using TF-IDF
vectorizer = TfidfVectorizer()
X_vectorized = vectorizer.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_vectorized, y, test_size=0.2, random_state=42)

# Genetic Algorithm parameters
POPULATION_SIZE = 1000
CROSSOVER_PROBABILITY = 0.5
MUTATION_PROBABILITY = 0.2
N_GENERATIONS = 50

# Create fitness and individual classes for the genetic algorithm
creator.create("FitnessMax", base.Fitness, weights=(1.0,))
creator.create("Individual", list, fitness=creator.FitnessMax)

# Define the toolbox for the genetic algorithm
toolbox = base.Toolbox()

# Register the necessary functions for the genetic algorithm
toolbox.register("attr_bool", np.random.choice, a=[True, False])
toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attr_bool, len(vectorizer.get_feature_names_out()))
toolbox.register("population", tools.initRepeat, list, toolbox.individual)
toolbox.register("evaluate", lambda individual: evaluate(individual, X_train, y_train, X_test, y_test))
toolbox.register("mate", tools.cxTwoPoint)
toolbox.register("mutate", tools.mutFlipBit, indpb=0.1)
toolbox.register("select", tools.selTournament, tournsize=3)

# Define the evaluation function for the SVM classifier
def evaluate(individual, X_train, y_train, X_test, y_test):
    # Select the features based on the individual's genome
    selected_features = [feature for feature, selected in zip(vectorizer.get_feature_names_out(), individual) if selected]
    X_train_selected = X_train[:, [vectorizer.vocabulary_.get(feature) for feature in selected_features]]
    X_test_selected = X_test[:, [vectorizer.vocabulary_.get(feature) for feature in selected_features]]

    # Train the SVM model
    svm_model = SVC()
    svm_model.fit(X_train_selected, y_train)

    # Evaluate the model
    y_pred = svm_model.predict(X_test_selected)
    fitness = np.mean(y_pred == y_test)

    return fitness,

# Run the genetic algorithm
population = toolbox.population(n=POPULATION_SIZE)
final_population, logbook = algorithms.eaSimple(population, toolbox, cxpb=CROSSOVER_PROBABILITY, mutpb=MUTATION_PROBABILITY, ngen=N_GENERATIONS)

# Get the best individual from the final population
best_individual = tools.selBest(final_population, k=1)[0]

# Select the features based on the best individual's genome
selected_features = [feature for feature, selected in zip(vectorizer.get_feature_names_out(), best_individual) if selected]
X_selected = X_vectorized[:, [vectorizer.vocabulary_.get(feature) for feature in selected_features]]

# Train the SVM model using the selected features
svm_model = SVC()
svm_model.fit(X_selected, y)

# Example sentence for prediction
new_complaint = "Loan servicing and payments problm with escrow account"

# Vectorize the new complaint using the selected features
new_complaint_vectorized = vectorizer.transform([new_complaint])
new_complaint_selected = new_complaint_vectorized[:, [vectorizer.vocabulary_.get(feature) for feature in selected_features]]

# Predict the class of the new complaint
predicted_class = svm_model.predict(new_complaint_selected)
print(f"The new complaint belongs to the class: {predicted_class}")
```
